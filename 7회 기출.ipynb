```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/happy2/happiness.csv',encoding='utf-8')
df.head()
```

```python
#각 년도의 행복랭킹 10위를 차지한 나라의 행복점수 평균
# display(df)
result = df[df['행복랭킹'] == 10]['점수'].mean()
print(result)
```

```python
# 각 년도의 행복랭킹 50위 이내의 나라
result = df[df['행복랭킹'] <= 50].sort_values('나라명')
# print(result)
result2 = result.groupby('나라명')['점수'].mean().to_frame()
print(result2)
#각각 행복점수 평균을 데이터프레임으로 표시
```

```python
# 2018년도 데이터들만 추출
df_1 = df[df['년도'] == 2018]
# print(df_1)
# 행복점수와 부패에 대한 인식에 대한 상관계수를 구하여라
result = df_1[['점수','부패에 대한인식']].corr().iloc[0,1]
print(result)
```

```python
#2018, 2019년도의 행복랭킹이 변화하지 않은 나라명의 수를 구하여라
result = len(df[['나라명','행복랭킹']]) - len(df[['나라명','행복랭킹']].drop_duplicates())
print(result)
```

```python
# 2019년도 데이터만 가져오기
# 각변수간 상관계수를 구하고 내림차순으로 정렬후 상위 5개, 데이터프레임으로 출력

result = df[df['년도'] == 2019].corr().unstack().to_frame().reset_index().dropna()
print(result)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/consum/Tetuan%20City%20power%20consumption.csv')
df.head()
```

```python
#datetime컬럼을 통해 각 월별로 몇개의 데이터가 있는지
df['DateTime'] = pd.to_datetime(df['DateTime'])
result = df['DateTime'].dt.month.value_counts().sort_index().to_frame()
# print(result)
print(df.info())
```

```python
# 3월달 각 시간대별 온도의 평균 구하기
month = df[df['DateTime'].dt.month == 3]
result = month.groupby(month.DateTime.dt.hour)['Temperature'].mean().sort_values().values[0]
print(result)
# 가장 낮은 시간대의 온도
```

```python
# Zone 1 Power Consumption 컬럼의 value값의 크기가 Zone 2 Power Consumption 컬럼의 value값의 크기보다 큰 데이터들의 Humidity의 평균을 구하여라
result = df[df['Zone 1 Power Consumption'] > df['Zone 2  Power Consumption']].Humidity.mean()
print(result)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p1.csv')
df.head(5)
```

국어,수학,영어,과학 과목 중 가장 많은 학생들이 응시한 시험을 선택하고 해당과목의 점수를 표준화 했을 때 가장 큰 표준화 점수를 구하여라

```python
# 가장 많은 학생들이 응시한 시험
test = df.isnull().sum()
# print(test) # 국어
# 해당과목의 점수를 표준화했을때 가장 큰 표준화 점수
import numpy as np
# ddof() : 모표준편차
result = ((df['국어'] - df['국어'].mean())/ df['국어'].std(ddof = 0)).max()
print(result)
```

**var_6 컬럼의 1,3사분위수 각각 IQR의 1.5배 벗어난 이상치의 숫자를 구하라 **

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3.csv')
df.head(5)
```

```python
IQR_1 = df['var_6'].quantile(0.25)
IQR_3 = df['var_6'].quantile(0.75)
IQR = IQR_3 - IQR_1

min = IQR_1 - 1.5 * IQR
max = IQR_3 + 1.5 * IQR
# print(IQR_1)
# print(IQR_3)
print(df[(df['var_6'] < min) | (df['var_6'] > max)])
```

```python
# 종속변수 :이용금액 , 평가지표 : rmse
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

# print(train.info())
# print(test.info()) # 이용금액이 종속변수

# 라벨인코딩
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

train['ID'] = le.fit_transform(train['ID'])
train['업종명'] = le.fit_transform(train['업종명'])
train['이용자구분'] = le.fit_transform(train['이용자구분'])
train['성별'] = le.fit_transform(train['성별'])

test['ID'] = le.fit_transform(test['ID'])
test['업종명'] = le.fit_transform(test['업종명'])
test['이용자구분'] = le.fit_transform(test['이용자구분'])
test['성별'] = le.fit_transform(test['성별'])

from sklearn.model_selection import train_test_split
x = train.drop(columns = ['ID','이용금액'])
y = train['이용금액']

X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2,random_state = 42)

from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(random_state = 1)
rfr.fit(X_train, Y_train)
pred = rfr.predict(X_test)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, pred)

import numpy as np
rmse = np.sqrt(mse)
# print(rmse)

#최종 결과 예측
test_x = test.drop(columns = ['ID'])
pred2 = rfr.predict(test_x)

pd.DataFrame({'이용금액 : ' : pred2}).to_csv('008003.csv', index = False)
print(pd.read_csv('008003.csv'))

# # pred_test = rfr.predict()


```
